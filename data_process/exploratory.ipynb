```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# busca data/ na raiz do projeto, independente de onde o kernel foi iniciado
_cwd = Path(os.getcwd())
DATA_DIR = (_cwd / "data") if (_cwd / "data").exists() else (_cwd.parent / "data")

L_VALUES = [8]  # [4, 8, 16, 32]
COLUMNS = ["i", "E", "lng", "H", "m", "m2", "m4"]

def load_dos(L):
    fpath = DATA_DIR / f"ising_DOS_L_{L}.dat"
    df = pd.read_csv(fpath, sep=r"\s+", comment="#", names=COLUMNS, engine="python")
    df["L"] = L
    return df

dfs = {}
for L in L_VALUES:
    fpath = DATA_DIR / f"ising_DOS_L_{L}.dat"
    if fpath.exists():
        dfs[L] = load_dos(L)
        print(f"L={L}: {len(dfs[L])} pontos carregados")
    else:
        print(f"L={L}: arquivo não encontrado ({fpath})")

if dfs:
    L0 = next(iter(dfs))
    display(dfs[L0].head(10))
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

colors = plt.cm.tab10(np.linspace(0, 0.9, max(len(dfs), 1)))

for (L, df), color in zip(dfs.items(), colors):
    ax.plot(df["E"], df["lng"], marker="o", ms=3, lw=1.2,
            color=color, label=f"L = {L}")

ax.set_xlabel(r"$", fontsize=13)
ax.set_ylabel(r"$\ln\,g(E)$", fontsize=13)
ax.set_title("Densidade de estados — Ising 2D (Wang-Landau)", fontsize=12)
ax.legend(frameon=False)
ax.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```
